# Import libraries

```python
import pandas as pd
import numpy as np
```

# Read train and test data

```python
df_train_x = pd.read_csv('../matrix_for_model/train_brand_matrix.csv',  index_col='device_id')
df_test_x = pd.read_csv('../matrix_for_model/test_brand_matrix.csv', index_col='device_id')

print(df_train_x.shape)
print(df_test_x.shape)
```

# Merge label_matrix

#### Read label_matrix data

```python
df_label = pd.read_csv('../matrix_for_model/label_matrix.csv',  index_col='device_id')
print(df_label.shape)
df_label.head()
```

#### Merge on device_id

```python
df_train_x = df_train_x.join(df_label, how='left')
print(df_train_x.shape)

df_test_x = df_test_x.join(df_label, how='left')
print(df_test_x.shape)
```

#### Fill in missing data

```python
df_train_x.fillna(0, inplace=True)
print('Missing value of df_train_x:', df_train_x.isnull().any().sum())

df_test_x.fillna(0, inplace=True)
print('Missing value of df_test_x',df_test_x.isnull().any().sum())
```

# Merge app_matrix

#### Read app_matrix data

```python
df_app = pd.read_csv('../matrix_for_model/app_matrix.csv',  index_col='device_id')
print(df_app.shape)
df_app.head()
```

#### Merge on device_id

```python
df_train_x = df_train_x.join(df_app, how='left')
print(df_train_x.shape)

df_test_x = df_test_x.join(df_app, how='left')
print(df_test_x.shape)
```

#### Fill in missing data

```python
df_train_x.fillna(0, inplace=True)
print('Missing value of df_train_x:', df_train_x.isnull().any().sum())

df_test_x.fillna(0, inplace=True)
print('Missing value of df_test_x',df_test_x.isnull().any().sum())
```

# Write df_train_x and df_test_x to CSV file

```python
df_train_x.to_csv('../matrix_for_model/train_x.csv')
df_test_x.to_csv('../matrix_for_model/test_x.csv')
```
